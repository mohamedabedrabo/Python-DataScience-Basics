```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0,10,1000)
y = np.sin(x)
```

```python
# plot :
plt.plot(x,y)
```

```python
# scatter plot :
plt.scatter(x,y)
```

```python
# or we can just take few points :
plt.scatter(x[::10],y[::10], color='red')
```
